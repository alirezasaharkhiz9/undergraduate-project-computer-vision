## car License plate detection
using CascadeClassifier

```python
cascade = cv2.CascadeClassifier('/content/drive/MyDrive/Colab Notebooks/DataSet/DATA/haarcascades/haarcascade_russian_plate_number.xml')

img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/DataSet/DATA/gauteng-new-number-plates-2023.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

car_plate = cascade.detectMultiScale(img_gray,scaleFactor=1.2,
    minNeighbors = 5, minSize=(25,25))

for (x,y,w,h) in car_plate:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),4)
    plate = img[y: y+h, x:x+w]

plt.imshow(img)
plt.show()
plt.imshow(plate)
```

## blurring a car license plate

```python
cascade = cv2.CascadeClassifier('/content/drive/MyDrive/Colab Notebooks/DataSet/DATA/haarcascades/haarcascade_russian_plate_number.xml')

img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/DataSet/DATA/gauteng-new-number-plates-2023.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

car_plate = cascade.detectMultiScale(img_gray,scaleFactor=1.2,
    minNeighbors = 5, minSize=(25,25))

for (x,y,w,h) in car_plate:
    # cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),4)
    plate = img[y: y+h, x:x+w]
    plate_blur = cv2.blur(plate,(11,11))
    img[y: y+h, x:x+w] = plate_blur

plt.imshow(img)
plt.show()
plt.imshow(plate)
```

## OCR with pyTesseract

#### The `pytesseract` library is a Python wrapper for the Tesseract Optical Character Recognition (OCR) engine. It is used to recognize and extract text from images. In summary, its working mechanism is as follows:

1. **Image Input**: An image containing text is passed to the `pytesseract` library. This image can include handwritten or printed text.

2. **Image Preprocessing**: Tesseract typically performs several preprocessing steps before processing the image, such as adjusting contrast and reducing noise. In some cases, users can manually perform these preprocessing steps to improve input quality.

3. **Segmenting the Image into Regions**: Tesseract divides the image into different regions, such as text areas, images, and graphics. These regions are identified based on the structure and features of the image.

4. **Character Recognition**: After identifying the text regions, Tesseract attempts to recognize characters (letters, numbers, and symbols) from the images pixels. This is done using machine learning algorithms based on large datasets and model training.

5. **Text Post-Processing**: After character recognition, Tesseract performs post-processing operations to improve the accuracy of the output, such as removing invalid characters and checking language dictionaries.

6. **Converting to Text**: Finally, Tesseract returns the recognized text as a string, which can be used in other applications.

With `pytesseract`, you can directly leverage Tesseracts OCR capabilities in Python. You can also specify additional settings like choosing the OCR language and configuring specific input/output formats.

In general, the mechanism of `pytesseract` works by utilizing Tesseracts OCR engine to identify and extract the text contained in images.

## convert car license plate to text

```python
cascade = cv2.CascadeClassifier('/content/drive/MyDrive/Colab Notebooks/DataSet/DATA/haarcascades/haarcascade_russian_plate_number.xml')

img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/DataSet/DATA/gauteng-new-number-plates-2023.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

car_plate = cascade.detectMultiScale(img_gray,scaleFactor=1.2,
    minNeighbors = 5, minSize=(25,25))

for (x,y,w,h) in car_plate:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),4)
    plate = img[y: y+h, x:x+w]

plt.imshow(img)
plt.show()
plt.imshow(plate)
plt.show()
```

```python
# image processing
plate = cv2.cvtColor(plate, cv2.COLOR_RGB2GRAY)
plate = cv2.GaussianBlur(plate,(3, 3),0)
plate = cv2.threshold(plate, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]
# convert with pytesseract
result = pytesseract.image_to_string(plate)
print(result)
```

```python
# convert with easyocr
reader = Reader(['en'])
result = reader.readtext(plate)
result
```
