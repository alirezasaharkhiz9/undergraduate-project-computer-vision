```python
from ultralytics import YOLO

model = YOLO('yolo11n-seg.pt') 

path = '/teamspace/studios/this_studio/tumor seg/Glioblastoma-Early-Diagnosis-2/data.yaml'
model.train(
    data=path,
    epochs=80,
    device=0,
)
```

```python
metrics = model.val()
print(metrics)
```

```python
pred = model.predict(
    source='/teamspace/studios/this_studio/tumor seg/Glioblastoma-Early-Diagnosis-2/test/images',
    save=True
)
print(pred)
```
